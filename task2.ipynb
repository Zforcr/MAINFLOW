```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
data = pd.read_csv('INDIAvi.csv')

```

```python
print(data.head())
print(data.describe())  # Summary statistics
print(data.info())  # Data types and missing values
print(data.isnull().sum())  # Check for missing values

```

```python
data[['views', 'likes', 'dislikes', 'comment_count']].hist(figsize=(10, 8), bins=30, color='skyblue')
plt.tight_layout()
plt.show()
```

```python
# Scatter Plot (Views vs Likes)
plt.figure(figsize=(8, 6))
sns.scatterplot(x='views', y='likes', data=data)
plt.title('Views vs Likes')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='views', y='dislikes', data=data)
plt.title('Views vs dislikes')
plt.show()
```

```python
# Box Plot (Likes by Category)
plt.figure(figsize=(12, 6))
sns.boxplot(x='category_id', y='likes', data=data)
plt.title('Likes Distribution by Category')
plt.show()
```

```python
# 3. Correlation Matrix Heatmap
plt.figure(figsize=(10, 8))
corr_matrix = data[['views', 'likes', 'dislikes', 'comment_count']].corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```

Based on the initial EDA findings, here are some hypotheses that can be developed for further analysis:

Hypotheses:
1.Videos with more views tend to receive more likes and comments.
2.Certain video categories (e.g., entertainment, music) tend to receive more likes and engagement compared to others.
3.Videos with disabled comments or ratings tend to have lower engagement.
4.Highly viewed videos attract more dislikes, possibly indicating more divisive content.
