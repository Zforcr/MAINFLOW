```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Extended sample data
data = {
    'Category': ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'J'],
    'Values': [23, 17, 35, 29, 12, 41, 30, 22, 18, 27],
    'Growth': [5, 7, 6, 9, 4, 10, 8, 3, 6, 7]
}

df = pd.DataFrame(data)
```

```python
# Create a bar chart for 'Category' vs 'Values'
plt.figure(figsize=(10, 6))
plt.bar(df['Category'], df['Values'], color='cornflowerblue')

# Add labels and title
plt.xlabel('Category')
plt.ylabel('Values')
plt.title('Bar Chart of Category vs Values')

# Show the plot
plt.show()
```

```python
# Create a line chart for 'Category' vs 'Growth'
plt.figure(figsize=(10, 6))
plt.plot(df['Category'], df['Growth'], marker='o', linestyle='-', color='orange', label='Growth Rate')

# Add labels, title, and legend
plt.xlabel('Category')
plt.ylabel('Growth Rate')
plt.title('Line Chart of Category vs Growth Rate')
plt.legend()

# Show the plot
plt.show()
```
