```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('realistic_large_dataset.csv')


# Display the first few rows of the dataset
print(df.head())

# Summary statistics of the dataset
print(df.describe())

# Check for missing values
print(df.isnull().sum())

# Visualization 1: Distribution of each variable (histograms)
df.hist(bins=10, figsize=(15, 10))
plt.suptitle('Distribution of Variables')
plt.show()

# Visualization 2: Box plots to identify outliers for each numerical feature
plt.figure(figsize=(10, 6))
sns.boxplot(data=df[['Age', 'Income', 'Spending Score', 'Purchase Frequency']], orient='h')
plt.title('Box Plot of Variables (Outlier Detection)')
plt.show()

# Visualization 3: Correlation heatmap to check for relationships between variables
plt.figure(figsize=(8, 6))
corr_matrix = df[['Age', 'Income', 'Spending Score', 'Purchase Frequency']].corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()

# Visualization 4: Scatter plot to visualize relationships between Age and Spending Score
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Age', y='Spending Score', data=df, hue='Gender', style='Gender', s=100)
plt.title('Scatter Plot between Age and Spending Score')
plt.show()

# Optional: Pair plot for visualizing pairwise relationships between numerical variables
sns.pairplot(df[['Age', 'Income', 'Spending Score', 'Purchase Frequency']])
plt.suptitle('Pair Plot of Variables', y=1.02)
plt.show()
```
